```python
import pandas as pd
import plotly.graph_objects as go

# 1. Leggi il file CSV
df = pd.read_csv("../1920_1929_coord.csv")

# 2. Raggruppa i dati per stato e conta il numero di attori per stato
actors_per_state = df.groupby('placeOfBirthLabel').size().reset_index(name='actorCount')

# 3. Prepara i dati necessari per la Choropleth map
# Assicurati che il DataFrame abbia una colonna con i nomi degli stati (puoi fare il mapping se necessario)
# Esempio:
# actors_per_state = actors_per_state.replace({"Stato1": "NomeStato1", "Stato2": "NomeStato2", ...})

# 4. Crea la Choropleth map utilizzando Plotly
fig = go.Figure(go.Choropleth(
    locationmode = 'country names',
    locations = actors_per_state['placeOfBirthLabel'],
    z = actors_per_state['actorCount'],
    text = actors_per_state['placeOfBirthLabel'],
    colorscale = 'Plasma',
    colorbar_title = 'Numero di attori',
))

# Aggiungi suddivisione degli Stati Uniti
fig.update_geos(
    resolution=50,
    showcountries=True, countrycolor="Black",
    showsubunits=True, subunitcolor="Blue"
)

fig.update_layout(
    title_text = 'Numero di attori per stato di nascita',
    geo=dict(
        showcoastlines=True,
    )
)

fig.show()
```
