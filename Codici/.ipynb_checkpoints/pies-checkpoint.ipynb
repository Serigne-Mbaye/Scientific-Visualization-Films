```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mpld3  # Import mpld3 library

# Assuming 'movies.csv' is the name of your CSV file
file_path = '../movies_r.csv'
df = pd.read_csv(file_path)

# Extract only the first genre from the 'Genre' column
df['FirstGenre'] = df['Genre'].str.split(',').str[0].str.strip()

# Scatter Plot - Relationship between ratings and meta scores with mpld3
fig, ax = plt.subplots(figsize=(10, 6))
scatter = sns.scatterplot(x='Rating', y='Meta Score', data=df, ax=ax)
mpld3.display(fig)
plt.figure(figsize=(10, 10))
# Identify genres with percentages under 5%
genre_counts = df['FirstGenre'].value_counts()
total_movies = len(df)
genre_percentages = genre_counts / total_movies * 100
under_5_percent_genres = genre_percentages[genre_percentages < 4].index

# Replace those genres with 'Other' in a new 'GroupedGenre' column
df['GroupedGenre'] = df['FirstGenre'].apply(lambda x: 'Other' if x in under_5_percent_genres else x)

# Scatter Plot - Relationship between ratings and meta scores with mpld3
fig, ax = plt.subplots(figsize=(10, 6))
scatter = sns.scatterplot(x='Rating', y='Meta Score', data=df, ax=ax)
mpld3.display(fig)
# Adjusting the explode parameter to separate slices
explode = [0.1 if i == df['GroupedGenre'].nunique() - 1 else 0 for i in range(df['GroupedGenre'].nunique())]

grouped_genre_counts = df['GroupedGenre'].value_counts()

# Displaying labels only for slices with a percentage greater than a threshold
threshold_percentage = 2.0  # Adjust as needed
visible_labels = grouped_genre_counts.index[grouped_genre_counts / total_movies * 100 >= threshold_percentage]

# Pie chart creation with legend labels
patches, texts, autotexts = plt.pie(
    grouped_genre_counts,
    labels=None,  # Labels will be set separately
    autopct=lambda p: '{:.1f}%'.format(p) if p > threshold_percentage else '',
    startangle=140,
    explode=explode,
)

# Set legend labels including "Other"
legend_labels = list(visible_labels) + ['Other']
plt.title('Percentage of Movies in Each Genre (Grouped)')
plt.legend(patches, legend_labels, loc='upper right', bbox_to_anchor=(1.2, 1))

# Display the pie chart
plt.show()
```
