```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import mpld3

# Caricamento dei dati
df = pd.read_csv('../Cinema_VS_Serie.csv', parse_dates=['Mese'], dayfirst=True)
df['Anno'] = df['Mese'].dt.year
df['Mese_num'] = df['Mese'].dt.month

# Pivot dei dati per la heatmap usando pivot_table
pivot_cinema = df.pivot_table(values='Cinema', index='Mese_num', columns='Anno')
pivot_serie = df.pivot_table(values='Serie televisiva', index='Mese_num', columns='Anno')

# Impostare il tema di Seaborn e il colore di sfondo
sns.set_theme(style="white")

# Creazione delle heatmap con sfondo bianco
fig, ax = plt.subplots(2, 1, figsize=(14, 14), facecolor='white')

# Heatmap per il cinema
sns.heatmap(pivot_cinema, cmap='YlGnBu', cbar=True, annot=True, fmt='g', ax=ax[0])
ax[0].set_title('Popolarità del Cinema', fontsize=16)
ax[0].set_xlabel('Anno', fontsize=12)
ax[0].set_ylabel('Mese', fontsize=12)

# Heatmap per le serie TV
sns.heatmap(pivot_serie, cmap='YlGnBu', cbar=True, annot=True, fmt='g', ax=ax[1])
ax[1].set_title('Popolarità delle Serie TV', fontsize=16)
ax[1].set_xlabel('Anno', fontsize=12)
ax[1].set_ylabel('Mese', fontsize=12)

plt.tight_layout()
plt.savefig("grafico_popolarita.png")
plt.show()
```
